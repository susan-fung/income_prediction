```python
# Import packages
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.feature_selection import SelectFromModel

#Classifiers
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

%matplotlib inline
```

```python
# Read in data
adult = pd.read_csv("data/adult.csv")
adult.head()
```

```python
# Check missing data for categorical data
for i in range(0,14):
    if adult.iloc[:,i].dtype == "O":
        print(adult.columns[i])
        print(sum(adult.iloc[:,i]=="?"))
```

```python
# Remove missing data
adult = adult[adult["workclass"] != "?"]
adult = adult[adult["occupation"] != "?"]
adult = adult[adult["native-country"] != "?"]
adult.shape
```

```python
adult.describe()
```

```python
# EDA
continuous = adult[["age","capital-gain","capital-loss","hours-per-week"]]
pd.plotting.scatter_matrix(continuous, figsize=(10,10));
```

```python
# One-hot-encode X
X = adult.iloc[:,0:14]
X = pd.get_dummies(X, columns = ["workclass", "education","marital-status",
                                  "occupation","relationship","race","gender","native-country"])
#X = X.as_matrix()
# Put column headings into an array
X_label = X.columns.values
```

```python
# 0 = less than or equal than 50K; 1: more than 50K
y = adult.iloc[:,14]
y = pd.get_dummies(y)
y = y.iloc[:,1]
y = y.as_matrix()

# Unbalanced classes in y
print("Number of observations less than or equal to 50K", sum(y == 0)/len(y))
print("Number of observations greater than 50K", 1-sum(y == 0)/len(y))
```

```python
# Feature selection
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1, stratify = y)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.20, random_state=1, stratify = y_train)
```

```python
# Model selection
classifiers = {
    'knn'           : KNeighborsClassifier(),
    'decision tree' : DecisionTreeClassifier(),
    'random forest' : RandomForestClassifier(),
    'log regression': LogisticRegression(),
    'naive Bayes'   : BernoulliNB() 
}

for c in classifiers:
    clf = classifiers[c]
    clf.fit(X_train, y_train)
    print(c)
    print("Training error: ", 1 - clf.score(X_train, y_train))
    print("Validation error: ", 1 - clf.score(X_val, y_val))
    print("\n")
```

```python
# Feature Selection
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
selected = SelectFromModel(rf,prefit = True)
X_selected = selected.transform(X_train)
print("Number of features selected:",X_selected.shape[1])
print("Selected features:")
print(X_train.columns[selected.get_support()])
```

```python
# Hyperparameters tuning
tuned_parameters = [{'n_estimators': list(range(1,10)),
                     'max_features': list(range(1,18)),
                    'max_depth': list(range(1,10))}]

clf = GridSearchCV(RandomForestClassifier(), tuned_parameters)
clf.fit(X_train, y_train)

print("Best parameters:")
print(clf.best_params_)
print()
print("Mean cross-validated error:")
print(1-clf.best_score_)
```
